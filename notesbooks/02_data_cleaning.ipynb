```python
import matplotlib.pyplot as plt

df['crash_hour'].value_counts().sort_index().plot(kind='bar')
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Hour of Day")
plt.show()
```

```python
pd.crosstab(
    df['weather_condition'],
    df['most_severe_injury']
).plot(kind='bar', stacked=True, figsize=(10,6))

plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.title("Injury Severity by Weather Condition")
plt.legend(title="Injury Severity", bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
pd.crosstab(
    df['lighting_condition'],
    df['most_severe_injury'],
    normalize='index'
).plot(kind='bar', stacked=True, figsize=(10,6))

plt.xlabel("Lighting Condition")
plt.ylabel("Proportion of Accidents")
plt.title("Injury Severity Distribution by Lighting Condition")
plt.legend(title="Injury Severity", bbox_to_anchor=(1.05, 1))
plt.show()
```

## Exploratory Data Analysis – Key Findings

1. Most traffic accidents occur during clear weather and daylight conditions.
   This is likely driven by higher traffic volume rather than increased risk.

2. Night-time and adverse weather conditions, while less frequent, show
   indications of higher injury severity, suggesting elevated per-incident risk.

3. Dry road surfaces account for most accidents due to exposure, but wet and
   icy surfaces present higher potential danger per trip.

4. Accident frequency peaks during typical commuting hours, highlighting the
   role of human behavior and congestion in accident occurrence.

5. Several categorical variables contain a significant number of "Unknown"
   values, indicating data collection limitations that must be addressed
   during cleaning and modeling.
